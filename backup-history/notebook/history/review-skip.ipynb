```python
import pandas
import numpy
from sklearn import metrics
```

```python
table = pandas.read_csv("./resource/20211117/alpha/csv/information.csv")
table['target'] = table['vote'].replace({"lower":0, 'higher':1})
table['prediction'] = table['prediction'].replace({"lower":0, 'higher':1})
print(table.shape)
table.head()
```

```python
print("-"*20, 'vote', "-"*20)
print(table['vote'].value_counts())
print("-"*20, 'mole_size', "-"*20)
print(table['mole_size'].value_counts())
print("-"*20, 'change', "-"*20)
print(table['change'].value_counts())
print("-"*20, 'gender', "-"*20)
print(table['gender'].value_counts())
print("-"*20, 'age', "-"*20)
print(table['age'].value_counts())
print("-"*20, "period", "-"*20)
print(table['period'].value_counts())
print("-"*20)
```

```python
def summary(table, label):

    score = table['score']
    prediction = table['prediction']
    target = table['target']
    mat = metrics.confusion_matrix(y_true=target, y_pred=prediction, labels=label)
    auc = metrics.roc_auc_score(y_score=score, y_true=target)
    tpr = mat[0,0] / sum(mat[0,:])
    tnr = mat[1,1] / sum(mat[1,:])
    ppv = mat[0,0] / sum(mat[:,0])
    npv = mat[1,1] / sum(mat[:,1])
    acc = numpy.diag(mat).sum() / mat.sum()
    output = {
        "matrix":numpy.round(mat,3),
        "acc":numpy.round(acc,3),
        "auc":numpy.round(auc,3),
        "tpr":numpy.round(tpr,3),
        "tnr":numpy.round(tnr,3),
        "ppv":numpy.round(ppv,3),
        "npv":numpy.round(npv,3)
    }    
    return(output)
```

```python
print("2021-11-17 before mole data and doctor label summary.")
summary(table=table, label=[1, 0])
```
